```python
from google.cloud import bigquery

client = bigquery.Client(project="valued-lambda-478823-h0")

table_id = "valued-lambda-478823-h0.azdd_test.uaf_voter_pref_counts_clean"

# grabs loaded csvs from google storage bucket
uri = "gs://azdd_test/uaf_voter_pref_counts_clean.csv"


job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,
    autodetect=True,
)

load_job = client.load_table_from_uri(uri, table_id, job_config=job_config)
load_job.result()

print("Loaded", client.get_table(table_id).num_rows, "rows into", table_id)
```
